# Tutorial 4: DataFrame
Un Dataframe es una versión extendida de una Serie, de hecho se podría decir que es un conjunto de series que actuan como una tabla excel o array multidimensional

```python
import pandas
```

```python
data = {"nombre":["Juan","Carlos","Marta","Lucas","Brais","Ana","Laura"],
        "puntuación":[60,45,30,65,100,50,87],
        "deporte":["Lucha","Futbol","Ballet","Surf","Esgrima","Gimnasia","Running"],
        "sexo":["M","M","F","M","M","F","F"]
       }
```

```python
dataFrame_personas = pandas.DataFrame(data) # Así formamos un dataframe a partir de datos de un diccionario de datos
```

```python
dataFrame_personas # Podemos visualizarlo como una tabla
```

```python
# Podemos reordenar las columnas en la creación del dataframe
dataFrame_personas = pandas.DataFrame(data, columns=["nombre", "deporte", "sexo", "puntuación"])
dataFrame_personas
```

```python
dataFrame_personas.head(3) #también contamos con las operaciones head y tail
```

```python
# Podemos añadir columnas nuevas y también cambiar los indices
dataFrame_personas = pandas.DataFrame(data,
                                      columns=["nombre", "deporte", "sexo", "puntuación", "edad"],
                                      index=["a","b","c","d","e","f","g"])
dataFrame_personas
```

```python
dataFrame_personas["deporte"] # Obtenemos un diccionario de deportes
```

```python
dataFrame_personas.deporte # Otra sintaxis posible
```

```python
# Obtenemos una parte del dataframe al usar de indice un array de las columnas que queremos
dataFrame_personas[["nombre","deporte"]]
```

```python
dataFrame_personas.loc[["a","f"]] # podemos obtener una fila o más por sus indice en un array con loc
```

```python
 # Podemos cambiar los datos de una columna con un array para cada fila o con un dato para que tengan todos el mismo
dataFrame_personas["edad"] =  [20,18,21,21,20,19,21]
dataFrame_personas
```

```python
# Podemos crear una columna nueva simplemente usandola para meter datos nuevos o con None

# Podemos hacer operaciones sobre los datos para obtener nuevos datos
dataFrame_personas["aprobado"] = dataFrame_personas.puntuación >= 50
dataFrame_personas
```

```python
del dataFrame_personas["aprobado"] # Borramos una columna con del
dataFrame_personas
```
